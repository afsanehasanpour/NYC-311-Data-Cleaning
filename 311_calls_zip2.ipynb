```python
import re#oral exam
def get_location_tuple(s):
    pt = re.compile(r'\-{0,1}\d{2}\.\d+')
    try:
        lst = re.findall(pt,s)
        lst = tuple(map(float,lst))
        return lst
    except:
        return np.nan
    
get_location_tuple(s)
```
